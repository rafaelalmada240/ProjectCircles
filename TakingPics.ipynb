```python
from cv2 import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
camera = VideoCapture(0)
while True:
    return_value,image = camera.read()
    gray = cvtColor(image,cv2.COLOR_BGR2GRAY)
    imshow('image',gray)
    if waitKey(1)& 0xFF == ord('s'):
        imwrite('test.jpg',image)
        break
camera.release()
destroyAllWindows()

# source from 
#https://stackoverflow.com/questions/11094481/capturing-a-single-image-from-my-webcam-in-java-or-python
```

```python
img = imread('test.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img[120:400,120:450])
```

```python
new_img = img[120:400,120:450]/np.max(img)
bin_img = new_img < np.max(new_img)/3
plt.imshow(bin_img)
```

```python
from skimage.morphology import skeletonize
skeleton = skeletonize(bin_img)
```

```python
plt.imshow(skeleton)
np.save('skeleton.npy',skeleton)
```
